```python
import contractions
import regex as re
```

```python
f = open('Skill_BOK (1).txt', 'r', encoding='utf-16')
data = f.read()
```

```python
data
```

```python
data = data.lower()

# remove references
data = re.sub("[[0-9]+]", "", data)

# remove arabic words
data = re.sub("[ا-ي]+", "", data)

# remove reduntant puctuations
characters = ['#', '-', '_', '$', '%', '[', ']',
              '"', '“', '”', '—', '§', '(', ')', '@', '/', '\\', '\n', "'"]

for i in characters:
    data = data.replace(i, ' ')

# remove numbers 
data = re.sub('[0-9]+', "", data)

# remove emails
data = re.sub('([a-zA-Z0-9._-]+@[a-zA-Z0-9._-]+\.[a-zA-Z0-9_-]+)', "", data)

# conract words
expanded_words = []
for word in data.split():
    # using contractions.fix to expand the shortened words
    expanded_words.append(contractions.fix(word))

    for i in range(0, len(expanded_words)-1):
        if expanded_words[i] == expanded_words[i+1]:
            expanded_words[i] = ''
data = ' '.join(expanded_words)

# handle punctuations
n = [' ', ',', '.', '=']
for i in n:
    reg = str("["+i+"]"+"{2,}")
    data = re.sub(reg, i, data)
```
